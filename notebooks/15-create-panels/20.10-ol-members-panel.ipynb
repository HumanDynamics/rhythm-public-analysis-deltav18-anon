# Load

# Master table

```python
members = pd.read_hdf(analysis_store_path, 'metadata/members')
members.head()
```

```python
members_master = members.copy()
```

```python
members_master['is_participant'] = 1
cond = (members_master.company == "Staff")
members_master.loc[cond,'is_participant'] = 0
```

```python
members_master
```

```python
members_master.dtypes
```

# Add entry survey info

```python
participants_entry_survey_data_clean = pd.read_hdf(surveys_clean_store_path, 'entry/participants_entry_survey_data_clean')
members_panel_entry_survey = participants_entry_survey_data_clean.copy()
```

# Add average daily subjective experience

```python
daily_survey_data_clean = pd.read_hdf(surveys_clean_store_path, 'daily/daily_survey_data_clean')
daily_survey_data_clean = daily_survey_data_clean.reset_index()
```

```python
members_panel_subjective = daily_survey_data_clean.groupby('member')[['Q1','Q2']].agg(['mean','count','min','max','var'])
members_panel_subjective.columns = ["_".join(x) for x in members_panel_subjective.columns.ravel()]
members_panel_subjective
```

# Merge with master panel

```python
members_panel = members_master.join(members_panel_entry_survey)
members_panel = members_panel.join(members_panel_subjective)
members_panel['is_ceo'] = members_panel['is_ceo'].fillna(0).astype(int)
members_panel['is_cofounder'] = members_panel['is_cofounder'].fillna(0).astype(int)
```

```python
members_panel.dtypes
```

# Store

```python
with pd.HDFStore(analysis_notebooks_store_path) as store:
    store.put('panels/members_panel',members_panel,format='table')
```
